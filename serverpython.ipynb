```python
#Create KNOWLEDGE BASE ::

bag_greet='hi hello morning evening hey'.upper().split()
bag_location="locations located situated place address ".upper().split()
bag_fee="fees amount fee installment".upper().split()
bag_stay='pg staying hostel room rooms flat flats accomodation'.upper().split()
bag_course='courses course program'.upper().split()
bag_cirriculum='cirriculum structure detail'.upper().split()
bag_timing='timing times duration '.upper().split()
bag_loan='loan'.upper().split()
bag_bloan='banks bank'.upper().split()
bag_discount='discount scholarship'.upper().split()
bag_exit='no thankyou bye byee exit close'.upper().split()
bag_about="about".upper().split()
bag_aegis="aegis muni".upper().split()
bag_self="you yourself your name".upper().split()
bag_event="event function events conference".upper().split()
bag_job="placement placements job jobs".upper().split()
bag_eligible="eligible eligibility apply reqirements qualifications qualification reqirement".upper().split()
bag_work="work experience".upper().split()
```

```python
### Creating Socket Connections for the Chat-bot Server
try:
    server=socket.socket()
    host=socket.gethostname()
    print("Connected on:",host)
    port=4444
    server.bind((host,port))
    print("Server done binding to host and port successfully")
    print("Waiting for Connections...")
    server.listen(1)
    client, addr= server.accept()
    print(addr,"Has connected to the server and is online")

    logfile, userinfo= [],[]
    datevisited=datetime.datetime.today().strftime("%d-%h-%Y")

    for i in range(0,3):
        name=client.recv(1024).decode()
        userinfo.append(name)

    while 1:
        rec_msg=client.recv(1024).decode()
        logfile.append(rec_msg)
        rec_msg=rec_msg.upper().split()


        eligible_counter=0
        work_counter=0
        course_counter=0
        job_counter=0
        event_counter=0
        aegis_counter=0
        self_counter=0
        intro_counter=0
        place_counter=0
        fee_counter=0
        stay_counter=0
        cstructure=0
        time_counter=0
        pna=0
        bloan=0
        isloan=0
        scholar=0

        for i in range(0,len(bag_location)):
            if bag_location[i] in rec_msg:
                place_counter+=1

        for i in range(0,len(bag_greet)):
            if bag_greet[i] in rec_msg:
                intro_counter+=1

        for i in range(0,len(bag_fee)):    
            if bag_fee[i] in rec_msg:
                fee_counter+=1

        for i in range(0,len(bag_course)):    
            if bag_course[i] in rec_msg:
                course_counter+=1

        for i in range(0,len(bag_stay)):    
            if bag_stay[i] in rec_msg:
                stay_counter+=1

        for i in range(0,len(bag_eligible)):    
            if bag_eligible[i] in rec_msg:
                eligible_counter+=1

        for i in range(0,len(bag_work)):    
            if bag_work[i] in rec_msg:
                work_counter+=1


        for i in range(0,len(bag_loan)):    
            if bag_loan[i] in rec_msg:
                isloan+=1

        for i in range(0,len(bag_event)):    
            if bag_event[i] in rec_msg:
                event_counter+=1

        for i in range(0,len(bag_job)):    
            if bag_job[i] in rec_msg:
                job_counter+=1

        for i in range(0,len(bag_discount)):    
            if bag_discount[i] in rec_msg:
                scholar+=1

        for i in range(0,5):
            for j in range (0,4):
                if (bag_location[i] in rec_msg) and (bag_fee[j] in rec_msg):
                    pna+=1

        for i in range(0,len(bag_course)):
            for j in range (0,len(bag_cirriculum)):
                if (bag_course[i] in rec_msg) and (bag_cirriculum[j] in rec_msg):
                    cstructure+=1

        for i in range(0,len(bag_course)):
            for j in range (0,len(bag_timing)):
                if (bag_course[i] in rec_msg) and (bag_timing[j] in rec_msg):
                    time_counter+=1

        for i in range(0,len(bag_about)):
            for j in range (0,len(bag_aegis)):
                if (bag_about[i] in rec_msg) and (bag_aegis[j] in rec_msg):
                    aegis_counter+=1

        for i in range(0,len(bag_about)):
            for j in range (0,len(bag_self)):
                if (bag_about[i] in rec_msg) and (bag_self[j] in rec_msg):
                    self_counter+=1

        for i in range(0,len(bag_loan)):
            for j in range (0,len(bag_bloan)):
                if (bag_loan[i] in rec_msg) and (bag_bloan[j] in rec_msg):
                    bloan+=1

        for i in range(0,len(bag_exit)):
            if bag_exit[i] in rec_msg:
                client.send("Thankyou !".encode())
                break

        if pna>0:
             client.send(fee_add.encode()) 

        elif cstructure>0:
            client.send(course_cirriculum.encode())

        elif aegis_counter>0:
            client.send(aegis_about.encode())

        elif self_counter>0:
            client.send(bot_about.encode())

        elif intro_counter>0:
            client.send(intro.encode())

        elif bloan>0:
            client.send(loan_bank.encode())

        elif scholar>0:
            client.send(scholarship.encode())

        elif time_counter>0:
            client.send(duration.encode())

        elif fee_counter>0:
            client.send(money.encode())

        elif stay_counter>0:
            client.send(living.encode())

        elif place_counter>0:
            client.send(adress.encode())

        elif job_counter>0:
            client.send(jobdetail.encode())

        elif course_counter>0:
            client.send(course.encode())

        elif eligible_counter>0:
            client.send(apply.encode())

        elif work_counter>0:
            client.send(exp.encode())

        elif event_counter>0:
            client.send(events.encode())

        elif isloan>0:
            client.send(loan_available.encode())

        else:
            if bag_exit[i] in rec_msg:
                client.close()
                break
            else:
                client.send('Query Not Understood, Please Try Again'.encode())
    print("\n",addr,"has disconnected the chat and is now Offline")

    querystr=''.join(logfile)
    uname=userinfo[0].title()
    uemail=userinfo[1]
    umobile=userinfo[2]

    log="visited.csv"
    file=open(log,"a")
    header="Name,Email,Phone,Date,Query\n"
    #file.write(header)
    file.write(uname + "," + uemail + "," + umobile + "," + datevisited + "," + querystr + "\n")
    file.close()
    print("\n\nServer Logs has been Generated :\n")
except:
    print("You need to restart the server.")
```
